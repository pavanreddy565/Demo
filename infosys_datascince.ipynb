```python
import pandas as pd
```

```python
df=pd.read_csv("rainfall1627650742214.csv")
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df.info
```

```python
#  if any and drop the corresponding rows.
df.dropna(inplace=True)
```

```python
len(df)
```

```python
df.columns
```

```python
# Find the district that gets the highest annual rainfall.
df.loc[df['ANNUAL'] == df['ANNUAL'].max(), "DISTRICT"]
```

```python
# Display the top 5 states that get the highest annual rainfall.
df1=df.sort_values(by=['ANNUAL'])
res=df1.groupby('STATE_UT_NAME').first()
res.iloc[0:5,0]
```

```python
# Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'.
df.drop([ 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'],axis=1)
```

```python
## Display the state-wise mean rainfall for all the months using a pivot table.

table=pd.pivot_table(df,values=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN',
       'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'ANNUAL'],index=['STATE_UT_NAME'],aggfunc="sum")
table.head()
```

```python
#Display the count of districts in each state.
df.groupby('STATE_UT_NAME').DISTRICT.count()
```

```python
# For each state, display the district that gets the highest rainfall in May. Also display the recorded rainfall.
# df.groupby('STATE_UT_NAME')['MAY'].max()
df.loc[df.groupby('STATE_UT_NAME')['MAY'].idxmax(),['STATE_UT_NAME', 'DISTRICT', 'MAY']]
```

```python
import numpy as np
x = np.arange(0.0, 5.0, 0.1)
y = np.cos(2 * np.pi * x) * np.exp(-x)
 
```

```python
!pip install matplotlib
```

# matplotlib targe graph


```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y, color='black', marker='o', markerfacecolor='lightblue', markersize=7)
plt.xlabel('Abscissa',fontsize=12,color='g')
plt.ylabel('Ordinate',fontsize=12,color='g')
plt.title('Main heading', fontsize=16, fontweight='bold')
plt.text(1, 0.45, '⌀=60°', va='center', rotation=0,
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3))
plt.annotate('2nd crest', xy=(2, 0.175), xytext=(2.5, 0.4), arrowprops=dict(facecolor='green', shrink=.01))
plt.show()
```
